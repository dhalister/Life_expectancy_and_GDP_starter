```python
#What is the average life expectancy in these nations? - LONGER SOLUTION
```

```python
life_span = life_expectancy['Life expectancy at birth (years)']
```

```python
mean_life_span_Chile = np.mean(life_span[life_expectancy.Country == 'Chile'])
```

```python
mean_life_span_China = np.mean(life_span[life_expectancy.Country == 'China'])
```

```python
mean_life_span_Germany = np.mean(life_span[life_expectancy.Country == 'Germany'])
```

```python
mean_life_span_Mexico = np.mean(life_span[life_expectancy.Country == 'Mexico'])
```

```python
mean_life_span_US = np.mean(life_span[life_expectancy.Country == 'United States of America'])
```

```python
mean_life_span_Zimbabwe = np.mean(life_span[life_expectancy.Country == 'Zimbabwe'])
```

```python
# test
print (mean_life_span_Zimbabwe)
```

```python
mean_life_span_by_country = [mean_life_span_Chile, mean_life_span_China, mean_life_span_Germany, mean_life_span_Mexico, mean_life_span_US, mean_life_span_Zimbabwe]
```

```python
country_list = ['Chile', 'China', 'Germany', 'Mexico', 'United States', 'Zimbabwe']
```

```python
plt.bar (country_list, mean_life_span_by_country)
ax = plt.subplot()
ax.set_xticks(['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe'])
ax.set_xticklabels(['Chile', 'China', 'Germany', 'Mexico', 'United States', 'Zimbabwe'], rotation = 30)
plt.xlabel("Countries")
plt.ylabel("Mean_life_expectancy_per_country")
plt.show()
plt.clf()
```

```python
#What is the distribution of that life expectancy? - shorter solution
```

```python
mean = life_expectancy.groupby('Country').mean('Life expectancy by birth (years)')
```

```python
print("Mean life expectancy: ")
print (mean)
```

```python
sns.barplot (x = 'Country', y = 'Life expectancy at birth (years)', data = mean)
ax = plt.subplot()
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(country_list, rotation = 30)
plt.show()
plt.clf()
```
